```python
import json

for n_qubits in range(3, 10):
    new_subgroups = {}
    subgroups = None
    with open(f"groups\\subgroup_generators_{n_qubits}.json", "r") as f:
        subgroups = json.load(f)
    for k, groups in subgroups.items():
        if k not in new_subgroups:
            new_subgroups[k] = []

        for generators in groups:
            new_generators = []
            for g in generators:
                new_generators.append(tuple([int(x) - 1 for x in g]))
            new_subgroups[k].append(new_generators)
    print(new_subgroups)
    with open(f"groups\\modified_subgroup_generators_{n_qubits}.json", "w") as f:
        json.dump(new_subgroups, f, indent=4)
```
